```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize']=8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
#Distribution:
vis1 = sns.distplot(stats["InternetUsers"])
```

```python
vis1 = sns.distplot(stats["InternetUsers"], bins=30)
```

```python
#Boxplots:
vis2 = sns.boxplot(data=stats, x="IncomeGroup", y="BirthRate")
```

```python
# ========== Seaborn Gallery ========#
```

```python
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', fit_reg=False )
```

```python
#Adding some color

vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', \
                  fit_reg=False, hue = 'IncomeGroup' )
```

```python
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', \
                  fit_reg=False, hue = 'IncomeGroup',size=10 )
```

```python
#====== Keyword Arguments in Python (Advanced Tutorial)
```

```python
#Marker size:

vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', \
                  fit_reg=False, hue = 'IncomeGroup',size=8, scatter_kws={"s":100} )
```
